```python
import pandas as pd
data = pd.read_csv("C:/Users/DELL/Desktop/Health_dataset.csv")
print(data)
```

```python
from datetime import datetime
data['Date of Birth'] = pd.to_datetime(data["Date of Birth"])
curr= pd.Timestamp.today().year
data["Age"]= curr-data["Date of Birth"].dt.year
print(data)
```

```python
bins = list(range(0, 111, 10))  # [0, 10, 20, 30, ..., 70, 80]
labels = [f"{i}-{i+9}" for i in bins[:-1]]  # ['0-9', '10-19', ..., '70-79']

# Categorize age into groups
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

age_group_counts = data['Age Group'].value_counts().reset_index()

# Rename columns for better readability
age_group_counts.columns = ['Age Group', 'Count']

# Sort the age groups in ascending order
age_group_counts['Age Group'] = pd.Categorical(age_group_counts['Age Group'], categories=labels, ordered=True)
age_group_counts = age_group_counts.sort_values('Age Group')


print(data)
# Display result
# print(data)
print(age_group_counts)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(age_group_counts['Age Group'], age_group_counts['Count'],  color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Group Distribution')
plt.xticks(rotation=45)
plt.show()
```
